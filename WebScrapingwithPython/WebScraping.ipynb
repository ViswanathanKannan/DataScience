## Scrape website to get stock data

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Quote Page
page= 'https://bloomberg.com/quote/SPX:IND'
```

```python
response = requests.get(page)
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
# Get the index name
name_box= soup.find('h1',attrs={'class':'name'})
```

```python
print(name_box.text.strip())
```

```python
# Get the index price
price_box = soup.find('div',attrs={'class':'price'})
```

```python
price_box.text
```
