```python
class Player(object):
    #class level Attribute
    
    def __init__(self, amount=100,betamount=10):
        self.amount=amount
        self.betamount = betamount
        
    def __str__(self):
        return("amount :" + str(self.amount))
    
    def won_bet(self):
        self.amount +=self.betamount
    
    def lost_bet(self):
        self.amount -= self.betamount
        
```

```python
import numpy as np
class Carddeck(object):
    
    #Class attribute
    deck = (("Ace-Spade",11),("King-Spade",10),("Queen-Spade",10),("Jack-Spade",10),("10-Spade",10),("9-Spade",9),("8-Spade",8),("7-Spade",7),("6-Spade",6),("5-Spade",5),("4-Spade",4),("3-Spade",3),("2-Spade",2)
            ,("Ace-Hearts",11),("King-Hearts",10),("Queen-Hearts",10),("Jack-Hearts",10),("10-Hearts",10),("9-Hearts",9),("8-Hearts",8),("7-Hearts",7),("6-Hearts",6),("5-Hearts",5),("4-Hearts",4),("3-Hearts",3),("2-Hearts",2)
            ,("Ace-Clubs",11),("King-Clubs",10),("Queen-Clubs",10),("Jack-Clubs",10),("10-Clubs",10),("9-Clubs",9),("8-Clubs",8),("7-Clubs",7),("6-Clubs",6),("5-Clubs",5),("4-Clubs",4),("3-Clubs",3),("2-Clubs",2)
            ,("Ace-Diamond",11),("King-Diamond",10),("Queen-Diamond",10),("Jack-Diamond",10),("10-Diamond",10),("9-Diamond",9),("8-Diamond",8),("7-Diamond",7),("6-Diamond",6),("5-Diamond",5),("4-Diamond",4),("3-Diamond",3),("2-Diamond",2))

    def __init__(self,pickedcards=[]):
        self.pickedcards = pickedcards

    def pickacard(self):
        while True:
            card = list(Carddeck.deck[np.random.randint(0,51)])
            if(card not in self.pickedcards):
                self.pickedcards.append(card)
                return card 
```

```python
from IPython.display import clear_output
player = Player(100,50)
carddeck = Carddeck()
dealercards=[]
playercards=[]

def printcardlist(cardlist):
    cardnumber = 1
    for card in cardlist:
        print("Card "+ str(cardnumber) + ": " + card[0])
        cardnumber+=1

def startagame():
    global player
    clear_output()
    if(player.amount<1):
        print("Amount reset to 100!!")
        player.amount=100
        
    print("Your current amount:"+ str(player.amount))
    while True:
        try:
            betamount = int(input("Bet amount: "))
        except:
            print('Please enter a valid amount in integer')
            continue
        else:
            if(betamount <= player.amount):
                player.betamount=betamount 
                break
            else:
                print('Bet amount is greater than your available balance')
            
    dealercards = [carddeck.pickacard()]
    playercards = [carddeck.pickacard()]
    dealercards.append(carddeck.pickacard())
    playercards.append(carddeck.pickacard())
    printcardlist(playercards)
    return dealercards,playercards

def checkposition(cardlist):
    value = 0
    for card in cardlist:
        value += int(card[1])
    return value

def checkwinning(dealercards,playercards):
    global player
    playervalue = checkposition(playercards)
    dealervalue = checkposition(dealercards)
    print("Player cards :"+str(playercards)+" Total:" + str(playervalue))
    print("Dealer cards :" +str(dealercards)+" Total:" + str(dealervalue))

    if(playervalue > 21):
        print('You got busted !!')
        player.lost_bet()
    elif(playervalue > dealervalue):
        print('Congrats. you have won!!')
        player.won_bet()
    elif(playervalue == dealervalue):
        print('It is a split!!')
    else:
        print('Dealer won!!')
        player.lost_bet()
   
        
def checkfornewgame():
        choice = input("To start another game. Press Y or y : ")
        if(choice.lower() == 'y'):
            return True
        else:
            return False

        
        
dealercards,playercards= startagame()
while True:
    choice = input("Please select your choice. Press H(hit) or S(stand)")
    if(choice.lower() == 'h'):
        checkwinning(dealercards,playercards)    
        if(checkfornewgame()):
            dealercards,playercards= startagame()
            pass
        else:
            break
        
    elif(choice.lower() == 's'):
        card =carddeck.pickacard()
        print("New card from deck : "+ card[0])
        playercards.append(card)
        if(checkposition(playercards) > 21):
            print("Player cards :"+str(playercards)+ " Total: " + str(checkposition(playercards)))
            print("Dealer cards :"+str(dealercards)+ " Total: " + str(checkposition(dealercards)))
            print('You got busted !!')
            player.lost_bet()
            if(checkfornewgame()):
                dealercards,playercards= startagame()
                pass
            else:
                break
    else:
        print('Correct choice was not provided')
        pass
    
```
